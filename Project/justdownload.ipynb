```python
import os
import json
from rich.progress import track

print(len(os.listdir("download/earth/")))
```

```python
def create_dl_links():
    dl_list =[]
    files = os.listdir("combined")
    for i in range(len(files)):
        with open("combined/" + files[i], "r") as file:
            data = json.load(file)
            for j in range(len(data)):
                date = data[j]["date"].split(" ")
                date.remove(date[-1])
                date = date[0].replace("-", "/")
                img = data[j]["image"] + ".png"
                link = f"https://epic.gsfc.nasa.gov/archive/enhanced/{date}/png/{img}"
                dl_list.append(link)
    return dl_list

list = create_dl_links()
print(len(list))
list.sort()
print(list[0])
```

```python
for i in track(range(1996, 3000)):
    os.system(f"wget -P download/earth/ {list[i]} --quiet")
    print(i)
```
